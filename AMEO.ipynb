## This is useful to perform all statistical techinques by column wise

```python
def univarient_analysis(numeric_data):
    for column in numeric_data:
        print("*" * 5, column, "*" * 5)
        print("min :", numeric_data[column].min())
        print("max :", numeric_data[column].max())
        print("mean :", numeric_data[column].mean())
        print("median :", numeric_data[column].median())
        print("IQR :",numeric_data[column].quantile())
              
        print("skewness :", numeric_data[column].skew())
        print("kurtosis :", numeric_data[column].kurt())
        print("---------------------------------------------------------------")

univarient_analysis(numeric)
```

## Skewness:
Skewness measures the asymmetry of the probability distribution of a random variable about its mean. It indicates whether the data is skewed to the left or right relative to the mean.

Negative Skewness: If the tail of the distribution extends more to the left of the mean, the distribution is negatively skewed.
Positive Skewness: If the tail of the distribution extends more to the right of the mean, the distribution is positively skewed.
Zero Skewness: A symmetric distribution has zero skewness.

## **Interpretation**

- **If skewness is close to 0, the distribution is approximately symmetric**.
- **If skewness is negative, the distribution is negatively skewed**.
- **If skewness is positive, the distribution is positively skewed**.

## Research Questions
-	Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you.
-	Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)


```python

job_roles = ["Programming Analyst", "Software Engineer", "Hardware Engineer", "Associate Engineer"]
filtered_data = df[df['Designation'].isin(job_roles)]


cs_graduates = filtered_data[filtered_data['Specialization'] == 'Computer Science']


avg_salary = cs_graduates['Salary'].mean()
print(f"Average Salary for Computer Science graduates in the given job roles: {avg_salary:.2f} lakhs")


if 2.5 <= avg_salary <= 3:
    print("The claim holds true based on the data.")
else:
    print("The claim does not hold true based on the data.")
```

```python

crosstab = pd.crosstab(df['Specialization'], df['Gender'])
print(crosstab)


from scipy.stats import chi2_contingency


chi2_stat, p_val, dof, ex = chi2_contingency(crosstab)


print(f"P-value from Chi-Square Test: {p_val:.4f}")


if p_val < 0.05:
    print("There is a significant relationship between gender and specialization.")
else:
    print("There is no significant relationship between gender and specialization.")
```
